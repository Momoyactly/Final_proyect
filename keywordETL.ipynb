```python
tweets = pd.read_sql( 'SELECT * FROM tweet',engine)
```

```python
def addKeywords(keywords):
    for word in keywords:
        tweets[word] = tweets['tweet'].str.find(word)
        tweets[word] = tweets[word].replace(-1, 1000)
    tweets[keywords[0]+'_words'] = tweets[keywords].min(axis=1)
    tweets.drop(keywords,axis=1,inplace=True)
```
